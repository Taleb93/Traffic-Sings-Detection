```python
import matplotlib.pyplot as plt

# Gegebene Werte
L = 0.5  # Kabellänge (Hinweg, 50 cm)
I = 9  # Stromstärke in Ampere
k = 58  # spezifischer Widerstand für Kupfer
U_input = 36  # Eingangsspannung in Volt

# Querschnitte
querschnitte = [0.75, 1, 1.5]  # Kabelquerschnitte in mm²
Ua_values = [(L * 2 * I) / (k * A) for A in querschnitte]  # Verlustspannungen (in Volt)
U_output = [U_input - Ua for Ua in Ua_values]  # Spannungen am Wandler

# Farben für die Punkte
colors = ['red', 'green', 'blue']

# Plot
plt.figure(figsize=(10, 6))

# Linie
plt.plot(querschnitte, U_output, linestyle='-', color='grey', label='Spannung am Wandler')

# Punkte mit Beschriftung
for i, (A, Ua, U, color) in enumerate(zip(querschnitte, Ua_values, U_output, colors)):
    plt.scatter(A, U, color=color, label=f"{A} mm²: Verlust: {Ua:.2f} V, Wandler: {U:.2f} V")
    plt.text(A, U - 0.5, f"{Ua:.2f} V Verlust\n{U:.2f} V Wandler", fontsize=10, ha='center')

# Achsenbeschriftungen
plt.xlabel("Kabelquerschnitt (mm²)", fontsize=12)
plt.ylabel("Spannung am Wandler (V)", fontsize=12)
plt.title("Spannung am DC-DC Wandler vs. Kabelquerschnitt", fontsize=14)

# Legende mit Zusatzinfo
plt.legend(loc='upper right', fontsize=10, title=f"Kabellänge: 50 cm\nStromstärke: {I:.2f} A")

# Achsenformatierung
plt.grid(True)
plt.axhline(U_input, color='grey', linestyle='--', label="Maximale Spannung (36 V)")
plt.xlim(0.5, 1.6)  # Begrenzung der X-Achse
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gegebene Werte
L = 0.5  # Kabellänge (Hinweg, 50 cm)
P = 30  # Leistung in Watt
U_input = 19  # Eingangsspannung in Volt
I = P / U_input  # Stromstärke in Ampere
k = 56  # spezifischer Widerstand für Kupfer

# Querschnitte
querschnitte = [0.75, 1, 1.5]  # Kabelquerschnitte in mm²
Ua_values = [(L * 2 * I) / (k * A) for A in querschnitte]  # Verlustspannungen (in Volt)
U_output = [U_input - Ua for Ua in Ua_values]  # Spannungen an der GPU

# Farben für die Punkte
colors = ['red', 'green', 'blue']

# Plot
plt.figure(figsize=(10, 6))

# Linie
plt.plot(querschnitte, U_output, linestyle='-', color='grey', label='Spannung an GPU')

# Punkte mit Beschriftung
for i, (A, Ua, U, color) in enumerate(zip(querschnitte, Ua_values, U_output, colors)):
    plt.scatter(A, U, color=color, label=f"{A} mm²: Verlust: {Ua:.2f} V, GPU: {U:.2f} V")
    plt.text(A, U - 0.5, f"{Ua:.2f} V Verlust\n{U:.2f} V GPU", fontsize=10, ha='center')

# Achsenbeschriftungen
plt.xlabel("Kabelquerschnitt (mm²)", fontsize=12)
plt.ylabel("Spannung an der GPU (V)", fontsize=12)
plt.title("Spannung an der GPU vs. Kabelquerschnitt", fontsize=14)

# Zusatzinfo zur Kabellänge und Stromstärke
plt.text(min(querschnitte), min(U_output) - 1, f"Kabellänge: 50 cm\nStromstärke: {I:.2f} A", 
         fontsize=12, color='blue', ha='left')

# Legende
plt.legend(loc='upper right', fontsize=10)

# Achsenformatierung
plt.grid(True)
plt.axhline(U_input, color='grey', linestyle='--', label="Maximale Spannung (19 V)")
plt.xlim(0.5, 1.6)  # Begrenzung der X-Achse
plt.show()
```
